```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dummy logs
df = pd.read_json("dummy_api_logs.json")  # Update with correct path if needed

# Convert timestamp to datetime for better analysis
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Set up the figure size for better visualization
plt.figure(figsize=(12, 6))

# Histogram of response times
sns.histplot(df["response_time_ms"], bins=30, kde=True, color="blue")
plt.axvline(x=2000, color='red', linestyle='dashed', label="Potential Anomaly Threshold (2000ms)")
plt.xlabel("Response Time (ms)")
plt.ylabel("Frequency")
plt.title("API Response Time Distribution")
plt.legend()
plt.show()

# Bar chart of status codes
plt.figure(figsize=(8, 5))
sns.countplot(x=df["status_code"], palette="coolwarm")
plt.xlabel("Status Code")
plt.ylabel("Count")
plt.title("API Response Status Code Distribution")
plt.show()

# Scatter plot: Response time over time
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="timestamp", y="response_time_ms", hue="status_code", palette="coolwarm", alpha=0.7)
plt.axhline(y=2000, color='red', linestyle='dashed', label="Anomaly Threshold (2000ms)")
plt.xlabel("Timestamp")
plt.ylabel("Response Time (ms)")
plt.title("API Response Time Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
